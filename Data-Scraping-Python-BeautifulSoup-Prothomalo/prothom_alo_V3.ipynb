```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import bs4
import csv
import re
```

```python
cnt = 1607842 #1610464
with open('prothom_alo_v3.csv', 'a', newline='', encoding='utf8') as fg:
    thewriter = csv.writer(fg)
    for p in range(100):
        cnt = cnt - 1
        string1 = "https://www.prothomalo.com/bangladesh/article/" + str(cnt) + '/'
        res = requests.get(string1)
        soup = bs4.BeautifulSoup(res.text, 'lxml')
        
        typ = soup.findAll("div", {"class": "breadcrumb"})
        #"div", {"class": "stylelistrow"}
        #print(typ)
        bt = 0
        for x in typ:
            lines = x.findAll('strong')
            pri = lines
            pri = str(pri)
            pri = re.sub('<strong.*?"title">','',pri, flags=re.DOTALL)
            pri = pri.replace('প্রচ্ছদ</strong>, ','')
            pri = pri.replace('</strong>','')
            pri = pri.replace('</strong>, ','')
            pri = pri.replace(', ','---')
            category = str(pri)
        result = soup.find(itemprop="datePublished")
        pub_date = str(result.text) + "/>"
        pub_date = re.sub(',.*?/>','',pub_date, flags=re.DOTALL)
        
        #print(pub_date)
        
        title = soup.find('h1').getText()
        #print(title)
        
        pri = soup.findAll(itemprop="articleBody")
#         pri = soup.findAll()
        for x in pri:
            lines = x.findAll('p')
            
        pri = lines
        pri = str(pri)
        pri = re.sub('<img.*?/>','',pri, flags=re.DOTALL)
        pri = pri.replace('<p>','')
        pri = pri.replace('</p>,','')
        pri = pri.replace('</p>','')
        pri = pri.replace('</br>','')
        pri = pri.replace('</br>','')
        pri = pri.replace('<p dir="','')
        pri = pri.replace('ltr','')
        pri = pri.replace('"">','')
        pri = pri.replace('<p class=""TEXT"">','')
        pri = pri.replace('<p dir=""ltr"">','')
        pri = pri.replace('"[ltr"">','')
       # "[ltr"">
        pri = pri.replace('  ',' ')
        pri = str(pri)
        
        news = pri
        print(cnt)
        pri = pri.replace('[','')
        pri = pri.replace(']','')
        if category == "[বাংলাদেশ---অপরাধ]":
            thewriter.writerow([cnt, pub_date, title, category, news])
```
